# **Factorial**

**Using Loop**

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
```

**Using Recursion**

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers."
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

**C++ Code**

```python
%%writefile fact.cpp

#include <iostream>
using namespace std;

int factorial(int n) {
    if (n == 1 || n == 0)
        return 1;
    else
        return n * factorial(n - 1);
}

int main() {
    int num = 5;
    cout << "Factorial of " << num << " is " << factorial(num) << endl;
    return 0;
}
```

# **Greatest Common Divisor (GCD)**

**The Euclidean Alogorithm**

```python
def find_gcd(a,b):
  while b != 0:
    a, b = b, a % b
  return a

a = 36
b = 12

print("GCD is:", find_gcd(a,b))
```

**Recursive Function For GCD**

```python
def gcd(a, b):
    if b == 0:
        return a
    else:
        return gcd(b, a % b)

a = 48
b = 18
print("GCD is:", gcd(a, b))
```

**C++ Code**

```python
%%writefile Rgcd.cpp

#include <iostream>
using namespace std;

int gcd(int a, int b) {
    if (b == 0)
        return a;
    else
        return gcd(b, a % b);
}

int main(){
    int a = 55;
    int b = 33;
    cout << "GCD is: " << gcd(a, b) << endl;
    return 0;
}
```

# **Fibonacci**

```python
def fibonacci(n):
    fibS = []
    a, b = 0, 1
    for _ in range(n):
        fibS.append(a)
        a, b = b, a + b
    return fibS

t = int(input("Enter the number of terms: "))
print(f"Fibonacci sequence up to {t} terms:")
print(fibonacci(t))
```

**C++ Code**

```python
%%writefile fibonacci.cpp
#include<iostream>
using namespace std;

void fibonacci(int n){
  int a = 0,b = 1,nxt;

  for(int i = 0; i < n; i++){
    cout << a << " ";
    nxt = a + b;
    a = b;
    b = nxt;
  }
  cout << endl;
}

int main(){
  int n;
  fibonacci(n);
  cout << "Enter the number of terms: ";
  cin >> n;
  cout << "Fibonacci sequence up to " << n << " terms:" << endl;
  fibonacci(n);
  return 0;
}
```
